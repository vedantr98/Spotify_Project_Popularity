```python
import os
import json
import requests

import pandas as pd

from dotenv import load_dotenv
from tqdm.notebook import tqdm
tqdm.pandas()

from IPython.display import Image
```

## 1. Spotify API setup

### 1.1 Loading Spotify API credentials from .env file

```python
load_dotenv()
```

```python
os.getenv("CLIENT_ID")
client_id=os.getenv("CLIENT_ID")
```

```python
client_secret=os.getenv("CLIENT_SECRET")
```

### 1.2 Obtaining Access Token 

```python
auth_url = 'https://accounts.spotify.com/api/token'
```

```python
# First we need to set up the credentials before sending the request
auth_url = 'https://accounts.spotify.com/api/token'

data = {
    'grant_type': 'client_credentials',
    'client_id': client_id,
    'client_secret': client_secret,
}
```

```python
auth_response = requests.post(auth_url, data=data)
```

```python
access_token = auth_response.json().get('access_token')
```

```python
access_token
```

```python
# We will create a variable to store access token
my_token=(access_token)
```

## 2. Collecting data 

Now that we have obtained the access token we can access and send a request for the required raw data

### 2.1 Artist Data

```python
# To collect the data we need to send a request 
url = "https://api.spotify.com/v1/artists?ids=6l3HvQ5sa6mXTsMTB19rO5%2C2YZyLoL8N0Wb9xBt1NhZWg"
headers = {
    "Authorization": f"Bearer {my_token}"
}
```

```python
response = requests.get(url, headers=headers)
```

```python
artists = response.json()
```

```python
# We will save this data to a JSON file
with open("../data/spotify_artists.json", "w") as f:
    json.dump(artists, f)
```
